Objectif : relier les donn√©es aux d√©cisions strat√©giques.

Analyses cl√©s :

Segments clients les plus rentables

Produits et sous-cat√©gories √† marges n√©gatives

Impact des remises sur les profits

Zones g√©ographiques √† forte/mauvaise rentabilit√©

Synth√®se du lien entre feedback client et performance

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donn√©es
data = pd.read_csv('C:/Users/georg/test/WCS_PYTON/Projets2026/Analyse_performances_commerciales/commandes_nettoyees.csv')

# 1Ô∏è‚É£ Segments clients les plus rentables
profit_segment = data.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
print(profit_segment)  # affichage des profits par segment

# Visualisation
sns.barplot(x=profit_segment.index, y=profit_segment.values)
plt.title('Profit par segment client')
plt.show()

# 2Ô∏è‚É£ Produits avec marges n√©gatives
negative_profit = data[data['Profit'] < 0] # filtrer commandes non rentables
top_negative_products = negative_profit.groupby('Product Name')['Profit'].sum().sort_values().head(10)
print(top_negative_products)  # top 10 produits d√©ficitaires

# 3Ô∏è‚É£ Impact des remises sur le profit par cat√©gorie
sns.boxplot(x='Category', y='Profit', data=data)
plt.title('Profit par cat√©gorie (impact des remises)')
plt.show()

# 4Ô∏è‚É£ Profits par r√©gion
profit_region = data.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(profit_region)

sns.barplot(x=profit_region.index, y=profit_region.values)
plt.title('Profit par r√©gion')
plt.show()

# 5Ô∏è‚É£ Feedback client vs profit
sns.boxplot(x='Feedback?', y='Profit', data=data)
plt.title('Impact du feedback client sur le profit')
plt.show()
```
